```python
import pandas as pd
import matplotlib.pyplot as plt

def plot_loss_vs_epoch(csv_file_path):
    # Read the data from the CSV file
    data = pd.read_csv(csv_file_path)

    # Extract columns for epoch, train loss, and validation loss
    epoch = data['Epoch']
    train_loss = data['Train Loss']
    val_loss = data['Val Loss']

    # Create the loss vs. epoch plot
    plt.figure(figsize=(10, 6))
    plt.plot(epoch, train_loss, label='Train Loss')
    plt.plot(epoch, val_loss, label='Validation Loss')
    plt.title('Loss vs. Epoch')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
# Usage:
csv_file_path = './att_unet_drive_200.csv'  # Replace with your CSV file path
plot_loss_vs_epoch(csv_file_path)
```
